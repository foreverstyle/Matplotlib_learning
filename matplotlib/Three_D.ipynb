```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# X, Y value
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)    # x-y 
R = np.sqrt(X ** 2 + Y ** 2)
# height value
Z = np.sin(R)

# 3D
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'), alpha=0.8)

# 
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.get_cmap('rainbow'), alpha=0.5)

# set the axis limits
ax.set_zlim(-2, 1)
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)


# # add contour
# ax.contour(X, Y, Z, zdir='z', offset=-2, cmap=plt.get_cmap('rainbow'))


# add contourf
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.get_cmap('rainbow'))

# show the plot
plt.show()
```
