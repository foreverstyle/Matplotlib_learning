```python
# %matplotlib notebook
# VS Code 和 Jupyter Notebook 有一些区别，特别是在处理交互式图形时。
# VS Code 支持通过内嵌 Jupyter Notebook 的界面来运行代码，但它的交互式显示功能可能不像在标准 Jupyter Notebook 中那样直接和流畅。
# 这可能导致动画或交互式图形无法正常显示或完全没有反应。

from matplotlib import pyplot as plt
from matplotlib import animation
import numpy as np

fig, ax = plt.subplots()

x = np.arange(0, 2*np.pi, 0.01)
line, = ax.plot(x, np.sin(x))


def animate(i):
    line.set_ydata(np.sin(x + i/10.0))
    return line,

def init():
    line.set_ydata(np.sin(x))
    return line,

ani = animation.FuncAnimation(fig=fig,
                              func=animate,
                              frames=100,
                              init_func=init,
                              interval=20,
                              blit=True)

# 1. fig 进行动画绘制的figure
# 2. func 自定义动画函数，即传入刚定义的函数animate
# 3. frames 动画长度，一次循环包含的帧数 
# 4. init_func 自定义开始帧，即传入刚定义的函数init 
# 5. interval 更新频率，以ms计 
# 6. blit 选择更新所有点，还是仅更新产生变化的点。应选择True，但mac用户请选择False，否则无法显示动画


# 你还可以将动画保存为 GIF 或视频文件，然后在笔记本中显示：
ani.save('sine_wave_animation.gif', writer='imagemagick')
# 确保你已经安装了 imagemagick 或其他相关的工具，以便能够保存和显示动画。

from IPython.display import Image
Image(filename="sine_wave_animation.gif")
```
